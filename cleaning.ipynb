```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("online_sales.csv")
df.head()
```

```python
df.tail()
```

```python
print(df.info())
print(df.shape)
print(df.describe())
print(df.columns)
df.columns = df.columns.str.strip()
```

```python
df.drop_duplicates()
```

```python
df.isnull().sum()
```

```python
# Handle CustomerID
df['CustomerID'].fillna('Annoymous', inplace=True)
# Handle WarehouseLocation 
df['WarehouseLocation'].fillna('Unknown', inplace=True)
# Set Shipping Cost to 0 where SalesChannel is "In-store" due to their beginning no shipping cost
df.loc[df['SalesChannel'] == 'In-store', 'ShippingCost'] = 0
```

```python
df.head(10)
```

```python
# Drop rows with NaN values:
df.dropna(inplace=True)
```

```python
df['InvoiceDate'] = pd.to_datetime(df['InvoiceDate'], errors = 'coerce')
```

```python
# List of columns to convert to categorical type
categorical_columns = ['PaymentMethod', 'Country', 'Category', 'SalesChannel', 'OrderPriority', 'ReturnStatus', 'ShipmentProvider', 'WarehouseLocation']

# Convert all listed columns to 'category' dtype
df[categorical_columns] = df[categorical_columns].apply(lambda x: x.astype('category'))
df.head()
```

```python
import numpy as np

# Calculate TotalPrice and round it up
df['TotalPrice'] = df.apply(
    lambda row: np.ceil(row['Quantity'] * row['UnitPrice']) if row['ReturnStatus'] != 'Returned' 
    else -np.ceil(row['Quantity'] * row['UnitPrice']), axis=1
)
```

```python
df.head(20)
```

```python
print(df.info())
print(df.describe())
```

```python
df = df[df['Quantity'] >= 0]

# Step 2: Adjust TotalPrice calculation based on ReturnStatus
df['TotalPrice'] = df.apply(
    lambda row: row['Quantity'] * row['UnitPrice'] if row['ReturnStatus'] != 'Returned' else -row['Quantity'] * row['UnitPrice'], axis=1
)
print(df.head())
```

```python
# Count the number of discounts greater than 1 without returns
discounts_above_1_no_returns = df[(df['Discount'] > 1) & (df['ReturnStatus'] != 'Returned')]
count_above_1_no_returns = len(discounts_above_1_no_returns)

print(f"Number of discounts greater than 1 without returns: {count_above_1_no_returns}")

# Check if all discounts > 1 are returns
all_discounts_above_1_are_returns = (df[df['Discount'] > 1]['ReturnStatus'].eq('Returned').all())
```

```python
# Save the cleaned DataFrame to a new CSV file
df.to_csv('cleaned_data.csv', index=False)

print("Cleaned data has been saved to 'cleaned_data.csv'.")
```
